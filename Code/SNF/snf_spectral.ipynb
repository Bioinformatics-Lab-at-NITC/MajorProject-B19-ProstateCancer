```python
clini['cluster'] = labels
clini
```

```python
death = []
time = []
duration = []
status = []
for days, followup, stat in zip(clini['days_to_death'].tolist(), clini['days_to_last_follow_up'].tolist(), clini['vital_status'].tolist()):
  if days == clini['days_to_death'].tolist()[0]:
    death.append(0)
    time.append(0)
  else: 
    death.append(1)
    time.append(days)

  if followup == clini['days_to_death'].tolist()[0]:
    duration.append(days)
  else: 
    duration.append(followup)
  if stat == "Alive": status.append(1)
  else: status.append(0)

clini['time'] = time
clini['death'] = death
clini['duration'] = duration
clini['status'] = status
clini
```
