```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#txt to csv
def txt_csv(path,filename):
  with open(path) as in_file:
    stripped = (line.strip() for line in in_file)
    lines = (line.split("\t") for line in stripped if line)
    with open(filename+'.csv', 'w') as out_file:
        writer = csv.writer(out_file)
        writer.writerows(lines)
txt_csv("/content/mrna_df.txt", 'meth')
mrna = pd.read_csv('meth.csv')
mrna_raw = mrna.dropna()
mrna_raw
```

```python
mrna = mrna_raw.transpose()
#meth = meth.iloc[:, 0:1000]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
mrna[list(mrna.columns.values)] = scaler.fit_transform(mrna[list(mrna.columns.values)])
mrna
```

```python
sample_type = []
column_names = mrna.index
for name in column_names[:]:
  sample_type.append(name[13:15])

mrna['sample type'] = sample_type
mrna = mrna.sort_values(by = 'sample type', ascending = 0)

mrna['sample type'] = mrna['sample type'].replace(['11'],0)
mrna['sample type'] = mrna['sample type'].replace(['01'],1)
mrna.reset_index(drop=True, inplace=True)
mrna.columns.name = None
mrna
```

```python
X = mrna.drop("sample type",1)   #Feature Matrix
y = mrna["sample type"] 
```
